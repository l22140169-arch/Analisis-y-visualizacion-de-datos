# EJERCICIO 4: Agrupación y Agregación

```python
import pandas as pd
df = pd.read_excel('ventas_mensuales.xlsx')
df['Total_Venta'] = df['Cantidad'] * df['Precio_Unitario']
resumen = df.groupby('Region')['Total_Venta'].sum()
print(resumen)
```

```python
promedio_por_producto = (
    df.groupby("Producto")["Cantidad"]
    .mean()
    .reset_index()
    .sort_values(by="Cantidad", ascending=False))

print(promedio_por_producto)
```

```python
import pandas as pd
df = pd.read_excel("empleados.xlsx")
salario_promedio = df.groupby("Departamento")["Salario"].mean()
print(salario_promedio)
```

```python
salarios_ciudad = df.groupby("Ciudad")["Salario"].agg(["min", "max"])
print(salarios_ciudad)
```

# Ejercicio 5: Limpieza de datos

```python
df_sucios = pd.read_excel('datos_con_problemas.xlsx')
df_sucios
```

```python
print(df_sucios.isnull().sum())
```

```python
df_sucios['Cliente'] = df_sucios['Cliente'].fillna('Cliente Desconocido')
df_sucios
```

```python
df_sucios = df_sucios.dropna(subset=['Monto'])
df_sucios
```

```python
df_sucios['Fecha'] = df_sucios['Fecha'].fillna('2024-01-01')
df_sucios
```

```python
df_sucios.to_excel('datos_limpios.xlsx', index=False)
print("Archivo guardado correctamente")
```

#EJERCICIO 7: Pipeline ETL Completo

```python
df = pd.read_excel('ventas_mensuales.xlsx')
df
```

```python
df['Total_Venta'] = df['Cantidad'] * df['Precio_Unitario']
df_filtrado = df[df['Total_Venta'] > 1000]
resumen = df_filtrado.groupby('Region')['Total_Venta'].sum().reset_index()
resumen = resumen.sort_values('Total_Venta', ascending=False)
```

```python
resumen.to_excel('reporte_regional.xlsx', index=False)
print('Pipeline completado exitosamente')
```
