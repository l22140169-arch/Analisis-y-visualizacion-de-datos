```r
#Required libraries
install.packages("readxl")
library(readxl)
```

```r
datos<-read_excel("data.xlsx")
head(datos)
```

#1.1.1 Media

```r
mean(datos$Estatura)
```

```r
apply(as.matrix(datos[,2:4]), 2,mean)
```

```r
lapply(datos[,2:4], mean)
```

```r
unlist(lapply(datos[,2:4], mean))
```

#1.1.2 Mediana

```r
median(datos$Estatura)
```

```r
apply(as.matrix(datos[,2:4]),2, median)
```

```r
unlist(lapply(datos[,2:4], median))
```

#1.2.1 Varianza y desviación estandar

```r
var(datos$Estatura)
```

```r
sd(datos$peso)
```

```r
apply(as.matrix(datos[,2:4]), 2, sd)
```

```r
unlist(lapply(datos[,2:4], sd))
```
